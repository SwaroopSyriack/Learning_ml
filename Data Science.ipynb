```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def scrappy(url):
    
    all = []
    
    headers = ({'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/54.0.2840.71 Safari/537.36'})
    webpage = requests.get(url,headers)
    soup = BeautifulSoup(webpage.content,"lxml")

    #The Title
    try:
        title = soup.find("span",attrs={"id":"productTitle"})
        title_value = title.string
        title_string = title_value.strip().replace(',', '')
    except AttributeError as e:
        title_string = "NA"
        print(e)
    
    print("product Title = ", title_string)
    all.append(title_string)
    
    # The price
    try:
        price = soup.find("div",attrs={"class":"a-section a-spacing-none aok-align-center aok-relative"})
        price = price.find("span").string.strip().replace(',','')
        
    except AttributeError as a:
        price = "NA"
        
    print("product price = ", price)
    all.append(price)

    #the ratings
    try:
        rating = soup.find("i",attrs={"class":"a-icon a-icon-star a-star-4-5 cm-cr-review-stars-spacing-big"}).string.strip().replace(',','')
    except AttributeError as a:
        try:
            rating = soup.find("span",attrs={"class":"a-icon-alt"}).string.strip().replace(',','')
        except:
            rating = "NA"       
    print("product Rating = ", rating)
    all.append(rating)

    #The avalibility
    try:
        available = soup.find("div", attrs={'id': 'availability'})
        available = available.find("span").string.strip().replace(',', '')
    except AttributeError as a:
        avalible = "NA"
        
    print("product Avalibility = ", avalible)
    all.append(avalible)
    return all
        
    
```

```python
if __name__ == "__main__":
    urls = open("C:/Users/swaroop/OneDrive/Desktop/urls.txt.txt","r")
    results = []
    for url in urls:
        scrap = scrappy(url)
        results.append(scrap)
```

```python
df = pd.DataFrame(results, columns=['item', 'Price', 'Rating','Avalibility'])
df.to_csv("C:/Users/swaroop/OneDrive/Desktop/out.csv", index=False)
```
