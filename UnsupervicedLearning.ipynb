```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
correlation_matrix = df.corr(method='pearson')  # or 'spearman'

threshold = 0.5
filtered_corr = correlation_matrix.where(abs(correlation_matrix) > threshold)

# Visualizing with a heatmap
plt.figure(figsize=(10,8))
sns.heatmap(filtered_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
from sklearn.feature_selection import VarianceThreshold

# Set a variance threshold (e.g., 0.1)
selector = VarianceThreshold(threshold=0.1)
selector.fit(df)

# Get the selected features
selected_features = df.columns[selector.get_support()]
print(selected_features)
```

```python
new_df = df[selected_features] 
```

```python
new_df
```

```python
correlation_matrix = new_df.corr(method='pearson')  # or 'spearman'

threshold = 0.5
filtered_corr = correlation_matrix.where(abs(correlation_matrix) > threshold)

# Visualizing with a heatmap
plt.figure(figsize=(10,8))
sns.heatmap(filtered_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```
